```python
import paho.mqtt.client as mqtt
import time
import random
import json

ACCESS_TOKEN = 
broker="demo.thingsboard.io"

lampadas_state = {1: False,2: False,3: False}
sensor_data = {"Sensor_1": 18,"Sensor_2": 18,"Sensor_3": 18}
temperature_rages = {"sala_1": {"min": 18, "max": 24},"sala_2": {"min": 18, "max": 24},"sala_3": {"min": 18, "max": 25} }

def on_connect(client, userdata, rc, extra_params):
    client.subscribe('v1/devices/me/rpc/request/+')

def on_message(client, userdata, message):

        print('Topico: ' + message.topic + '\nMensagem ' + str(message.payload))
        data = json.loads(message.payload)

        # controle de acionamento das lâmpadas
        if data['method'] == 'getGpioStatus':
          client.publish(message.topic.replace('request','response'), get_lampada_status(),1)
        elif data['method'] == 'setGpioStatus':
          set_lampada_status(data['params']['pin'], data['params']['enabled'])
          client.publish(message.topic.replace('request','response'), get_lampada_status(),1)
          client.publish('v1/devices/me/attributes', get_lampada_status(),1)
       

       # SALA 1:
       # controle dos valores atuais de temperatura nas sala
        elif data['method'] == 'getValue_1':
          client.publish(message.topic.replace('request','response'), get_temp_value_1(),1)
        elif data['method'] == 'setValue_1':
          set_temp_value_1(data['params'],message.topic)
          client.publish(message.topic.replace('request','response'), get_temp_value_1(),1)
          client.publish('v1/devices/me/attributes', get_temp_value_1(),1)

       # ---------------------------------------------------------------------------------- 

        elif data['method'] == 'getValueMin1':
          client.publish(message.topic.replace('request','response'), get_temp_min_1(),1)
        elif data['method'] == 'setValueMin1':
          set_temp_min_1(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_min_1(),1)
          client.publish('v1/devices/me/attributes', get_temp_min_1(),1)

        elif data['method'] == 'getValueMax1':
          client.publish(message.topic.replace('request','response'), get_temp_max_1(),1)
        elif data['method'] == 'setValueMax1':
          set_temp_min_1(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_max_1(),1)
          client.publish('v1/devices/me/attributes', get_temp_max_1(),1)
       # controle dos valores maximos e minimos de temperatura na sala
        
      # SALA 2:
      # controle dos valores atuais de temperatura nas sala

        elif data['method'] == 'getValue_2':
          client.publish(message.topic.replace('request','response'), get_temp_value_2(),1)
        elif data['method'] == 'setValue_2':
          set_temp_value_2(data['params'],message.topic)
          client.publish(message.topic.replace('request','response'), get_temp_value_2(),1)
          client.publish('v1/devices/me/attributes', get_temp_value_2(),1)

      # ---------------------------------------------------------------------------------- 

        elif data['method'] == 'getValueMin2':
          client.publish(message.topic.replace('request','response'), get_temp_min_2(),1)
        elif data['method'] == 'setValueMin2':
          set_temp_min_2(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_min_2(),1)
          client.publish('v1/devices/me/attributes', get_temp_min_2(),1)

        elif data['method'] == 'getValueMax2':
          client.publish(message.topic.replace('request','response'), get_temp_max_2(),1)
        elif data['method'] == 'setValueMax2':
          set_temp_max_2(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_max_2(),1)
          client.publish('v1/devices/me/attributes', get_temp_max_2(),1)
        
      # controle dos valores maximos e minimos de temperatura na sala

      # SALA 3:
      # controle dos valores atuais de temperatura nas sala
        elif data['method'] == 'getValue_3':
          client.publish(message.topic.replace('request','response'), get_temp_value_3(),1)
        elif data['method'] == 'setValue_3':
          set_temp_value_3(data['params'],message.topic)
          client.publish(message.topic.replace('request','response'), get_temp_value_3(),1)
          client.publish('v1/devices/me/attributes', get_temp_value_3(),1)

     # ---------------------------------------------------------------------------------- 

        elif data['method'] == 'getValueMin3':
          client.publish(message.topic.replace('request','response'), get_temp_min_3(),1)
        elif data['method'] == 'setValueMin3':
          set_temp_min_3(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_min_3(),1)
          client.publish('v1/devices/me/attributes', get_temp_min_3(),1)

        elif data['method'] == 'getValueMax3':
          client.publish(message.topic.replace('request','response'), get_temp_max_3(),1)
        elif data['method'] == 'setValueMax3':
          set_temp_max_3(data['params'])
          client.publish(message.topic.replace('request','response'), get_temp_max_3(),1)
          client.publish('v1/devices/me/attributes', get_temp_max_3(),1)

       # controle dos valores maximos e minimos de temperatura na sala


     
def get_lampada_status():
  return json.dumps(lampadas_state)

def set_lampada_status(pin,status):
  if status is None:
    lampadas_state[pin] = lampadas_state[pin]
  else:
    lampadas_state[pin] = status
  print('------------> Estado da Lâmpada <-------------- ')
  print(f'Lâmpada {pin} - ', 'Ligada' if lampadas_state[pin] else 'Desligada')


# ------------------------------------ SALA 1 ----------------------------------

def get_temp_value_1():
  return json.dumps(sensor_data)

def set_temp_value_1(value):
  if value is None:
    sensor_data["Sensor_1"] = sensor_data["Sensor_1"]
  else:
    if value < temperature_rages["sala_1"]["min"]:
      sensor_data["Sensor_1"] = temperature_rages["sala_1"]["min"]
    elif value > temperature_rages["sala_1"]["max"]:
      sensor_data["Sensor_1"] = temperature_rages["sala_1"]["max"]
    else:
      sensor_data["Sensor_1"] = value


#---------------------------------- VALORES MAXIMOS E MINIMOS ---------------------------

def get_temp_max_1():
  return json.dumps(temperature_rages["sala_1"]["max"])

def set_temp_max_1(value):
  if value is None:
    temperature_rages["sala_1"]["max"] = temperature_rages["sala_1"]["max"]
  else:
    temperature_rages["sala_1"]["max"] = value


def get_temp_min_1():
  return json.dumps(temperature_rages["sala_1"]["min"])

def set_temp_min_1(value):
  if value is None:
    temperature_rages["sala_1"]["min"] = temperature_rages["sala_1"]["min"]
  else:
    temperature_rages["sala_1"]["min"] = value


# ------------------------------------ SALA 2 ----------------------------------

def get_temp_value_2():
  return json.dumps(sensor_data)

def set_temp_value_2(value):
  if value is None:
    sensor_data["Sensor_2"] = sensor_data["Sensor_2"]
  else:
    if value < temperature_rages["sala_2"]["min"]:
      sensor_data["Sensor_2"] = temperature_rages["sala_2"]["min"]
      # client.publish(topic.replace('request','response'), get_temp_value_2(),1)
      client.publish('v1/devices/me/attributes', get_temp_value_2(),1)
    elif value > temperature_rages["sala_2"]["max"]:
      sensor_data["Sensor_2"] = temperature_rages["sala_2"]["max"]
      # client.publish(topic.replace('request','response'), get_temp_value_2(),1)
      client.publish('v1/devices/me/attributes', get_temp_value_2(),1)
    else:
      sensor_data["Sensor_2"] = value

def get_temp_value_2():
  return json.dumps(sensor_data)

#---------------------------------- VALORES MAXIMOS E MINIMOS ---------------------------

def get_temp_max_2():
  return json.dumps(temperature_rages["sala_2"]["max"])

def set_temp_max_2(value):
  if value is None:
    temperature_rages["sala_2"]["max"] = temperature_rages["sala_2"]["max"]
  else:
    temperature_rages["sala_2"]["max"] = value

def get_temp_min_2():
  return json.dumps(temperature_rages["sala_2"]["min"])

def set_temp_min_2(value):
  if value is None:
    temperature_rages["sala_2"]["min"] = temperature_rages["sala_2"]["min"]
  else:
    temperature_rages["sala_2"]["min"] = value



#------------------------------------ SALA 3 ----------------------------------
def get_temp_value_3():
  return json.dumps(sensor_data)

def set_temp_value_3(value):
  if value is None:
    sensor_data["Sensor_3"] = sensor_data["Sensor_3"]
  else:
    if value < temperature_rages["sala_3"]["min"]:
      sensor_data["Sensor_3"] = temperature_rages["sala_3"]["min"]
      # client.publish(topic.replace('request','response'), get_temp_value_3(),1)
      client.publish('v1/devices/me/attributes', get_temp_value_3(),1)
    elif value > temperature_rages["sala_3"]["max"]:
      sensor_data["Sensor_3"] = temperature_rages["sala_3"]["max"]
      # client.publish(topic.replace('request','response'), get_temp_value_3(),1)
      client.publish('v1/devices/me/attributes', get_temp_value_3(),1)
    else:
      sensor_data["Sensor_3"] = value

#---------------------------------- VALORES MAXIMOS E MINIMOS ---------------------------

def get_temp_max_3():
  return json.dumps(temperature_rages["sala_3"]["max"])

def set_temp_max_3(value):
  if value is None:
    temperature_rages["sala_3"]["max"] = temperature_rages["sala_3"]["max"]
  else:
    temperature_rages["sala_3"]["max"] = value


def get_temp_min_3():
  return json.dumps(temperature_rages["sala_3"]["min"])

def set_temp_min_3(value):
  if value is None:
     temperature_rages["sala_3"]["min"] = temperature_rages["sala_3"]["min"]
  else:
    temperature_rages["sala_3"]["min"] = value


client=mqtt.Client()
client.on_connect=on_connect
client.on_message= on_message
client.username_pw_set(ACCESS_TOKEN)
client.connect(broker,port=1883, keepalive=60)  
client.publish('v1/devices/me/attributes',get_lampada_status(),1)

# print("Conectando-se ao Broker: ",broker)

client.loop_start()

# try:
#   client.loop_forever()
# except KeyboardInterrupt:
#   print('stop')
#   client.loop_stop()

while True:
  try:
    # a = random.randint(18,24)
    # b = random.randint(18,24)
    # c = random.randint(18,25)
    # sensor_data['Sensor_1'] = a
    # sensor_data['Sensor_2'] = b
    # sensor_data['Sensor_3'] = c
    client.publish("v1/devices/me/telemetry",json.dumps(sensor_data),1)
    
  except KeyboardInterrupt:
    print('stop')
    client.loop_stop()
  time.sleep(1)
```
